```
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.mode.chained_assignment = None 

inc_data = pd.read_csv("incarceration_trends.csv")
```

```
inc_data["black_prison_pop_ratio"] = inc_data["black_prison_pop"]/inc_data["total_prison_pop"]
```

```
MN_data =  inc_data.loc[inc_data["state"] == "MN"]

MN_data_1 = MN_data.loc[(MN_data["year"] == 1977) | (MN_data["year"] == 1981) | (MN_data["year"] == 1985)]
MN_data_2 = MN_data.loc[(MN_data["year"] == 1989) | (MN_data["year"] == 1993) | (MN_data["year"] == 1997)]                
MN_data_3 = MN_data.loc[(MN_data["year"] == 2001) | (MN_data["year"] == 2005) | (MN_data["year"] == 2009)]


g = sns.FacetGrid(MN_data_1, col = "year", height = 10)
g.map(sns.kdeplot,"black_prison_pop", shade = True)
g.set(ylim=(0,.01))
j = sns.FacetGrid(MN_data_2, col = "year", height = 10)
j.map(sns.kdeplot,"black_prison_pop", shade = True)
j.set(ylim = (0,.01))
k = sns.FacetGrid(MN_data_3, col = "year", height = 10 )
k.map(sns.kdeplot,"black_prison_pop", shade = True)
k.set(ylim=(0,.01))

sns.despine(left = True, bottom = True)

plt.show()
```
